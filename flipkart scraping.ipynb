```python
from autoscraper import AutoScraper
```

```python
flipkart_url = "https://www.flipkart.com/search?q=iphone+13&sid=tyy%2C4io&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_2_6_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_2_6_na_na_na&as-pos=2&as-type=RECENT&suggestionId=iphone+13%7CMobiles&requestId=4b9b7eec-c3e6-4658-8af6-de8d19adef8a&as-searchtext=iphone"
wanted_list = ['₹69,999',"APPLE iPhone 13 (Starlight, 128 GB)","969 Reviews"]
```

```python
scraper = AutoScraper()
result = scraper.build(flipkart_url,wanted_list)
print(result)
```

```python
scraper.get_result_similar(flipkart_url,grouped=True)
```

```python
scraper.set_rule_aliases({'rule_u9yb':'title','rule_dypb': 'price','rule_ntym': 'reviews'})
scraper.keep_rules(['rule_u9yb','rule_dypb','rule_ntym'])
scraper.save('flipkart_search')
```
